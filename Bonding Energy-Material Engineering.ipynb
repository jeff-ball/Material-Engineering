# Finding ro  and Eo

$ E_{a} = \frac{-1.436}{r}$

$E_{R} = \frac{5.8}{r^9\times10^6}$

```
import numpy as np
r = np.zeros (100)
r

```

```
p = 0.01
for i in range (len(r)): 
  r[i] = p
  p = p + 0.005
```

```
r
```

```
import pandas as pd
table  = pd.DataFrame()
```

```
Ea = -1.436/r
Er = 5.8/(r**(9)*10**(6))
print (Ea,'\n',Er)
```

```
table['r'] = r
table ['Ea'] = Ea
table['Er']= Er
```

```
table
```

```
import matplotlib.pyplot as plt
#plotting Ea versurs r
plt.plot(r,Ea)
```

```
#plotting Er versus r
plt.plot(r,Er)
```

```
E = Ea + Er
table['E'] = E
```

```
table
```

```
plt.plot(r,E)
```

```
import seaborn as sns
sns.set()
plt.figure(figsize = (10,6))
plt.plot(r,Ea,'y--')
plt.plot(r,Er,'g--')
plt.plot(r,E)
plt.grid(True)
plt.ylim((-12,12))
plt.show()
```

# Some quick math using Scipy !!

```
# u can ignore this section and the curvefit for now/ jump to the cell below!
from scipy.misc import derivative 
def f(r): 
  return  ((5.8/((r**9)*(10**6)))-(1.436/r))

derivative (f, 0.2785353535353535, dx = 1e-6)
  
```

**Optimization algorithm using Scipy to find roots**

```
#from scipy import optimize 
#optimize.newton(f,0.26,tol=0.00001)
```

```
array = np.linspace(0.275,0.285,100)
array
```

```

for i in array :
  #print(i) to see progress
  # you can play with the array to find accurate values
  if (int(derivative (f, i, dx = 1e-6))  == 0):
    print (i)
    break
  else: 
    continue
```

```
for i in array :
  #print(i) to see progress
  # you can play with the array to find accurate values
  if -0.1 <= (derivative (f, i, dx = 1e-6)) <= 0.1:
    print (i)
    break
  else: 
    continue
```

**We got a more accurate value** <br>
